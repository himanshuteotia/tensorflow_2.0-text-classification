```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Flatten, Dense
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
dataset = pd.read_csv('Churn_Modelling.csv')
```

```python
print(# dataframe.size 
 "size of dataframe",dataset.size
)
print(# dataframe.shape 
 "shape of dataframe",dataset.shape
)
dataset.head()
```

```python
X = dataset.drop(labels=['CustomerId','Surname','RowNumber','Exited'],axis = 1) # here axis=0 represents rows and axis=1 represents columns
Y = dataset['Exited']
```

```python
X.head(20)
```

```python
Y.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label1 = LabelEncoder()
X['Geography'] = label1.fit_transform(X['Geography'])
```

```python
X.head(20)
```

```python
label = LabelEncoder()
X['Gender'] = label.fit_transform(X['Gender'])
```

```python
X.tail()
```

```python
X = pd.get_dummies(X, drop_first=True,columns=['Geography']) # one hot encoding using pandas we can also do this using sklearn
```

```python
X.head(20)
```

```python
#get all unique values from column
dataset.Geography.unique()
```

### Feature Standarization

```python
from sklearn.preprocessing import StandardScaler
```

```python
X_train,X_test, y_train, y_test = train_test_split(X,Y,test_size=0.2,random_state=0,stratify=Y)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

**Build ANN**



```python
#using shift+ double tab will show us the documentation of any function
model = Sequential()
print("Shape of X : ", X.shape[1]) # just to print the value
model.add(Dense(X.shape[1], activation='relu', input_dim=X.shape[1]))
model.add(Dense(128,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train,y_train.to_numpy(),batch_size=10, epochs=10, verbose=1)
```

```python
y_pred = model.predict_classes(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
model.evaluate(X_test,y_test.to_numpy())
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```
